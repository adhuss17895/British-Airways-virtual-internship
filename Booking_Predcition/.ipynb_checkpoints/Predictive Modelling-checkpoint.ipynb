```python
import pandas as pd
import numpy as np 
import os 

import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
cwd = os.getcwd()

df = pd.read_csv(cwd + "/new_customer_booking.csv", index_col=0)
df.head()
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df_final = df
df_final
```

```python
from sklearn.preprocessing import OneHotEncoder

#Create one instance of one hot encoder
encoder = OneHotEncoder(handle_unknown='ignore')

#one hot encode for sales-channel
encoder_sc = pd.DataFrame(encoder.fit_transform(df[["sales_channel"]]).toarray())
encoder_sc = encoder_sc.rename(columns={0:'Internet',1:'Mobile'})
df_final = df_final.join(encoder_sc)

encoder_tt = pd.DataFrame(encoder.fit_transform(df[["trip_type"]]).toarray())
encoder_tt = encoder_tt.rename(columns={0:'RoundTrip',1:'OnewWayTrip',2:'CircleTrip'})
df_final = df_final.join(encoder_tt)
                                        
```

```python
#Drop categorical columns
df_final.drop(["sales_channel","trip_type","booking_origin","route"], axis=1, inplace=True)
```

```python
#labelling the data for supervised learning
label = df["booking_complete"]
```

```python
#Drop the labelled data from final data frame
df_final = df_final.drop("booking_complete",axis=1)
```

```python
df_final.info()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#fit and trasnform dta
scaled_df = scaler.fit_transform(df_final)
```

```python
#create a dataframe of the scaled df
scaled_df = pd.DataFrame(scaled_df,columns=df_final.columns)
```

```python
#add the labelled column to scaled df
scaled_df['label'] = label
```

```python
scaled_df
```

```python
corr = scaled_df.corr()

plt.figure(figsize=(10,7))

#heatmap
sns.heatmap(corr)
```

```python
#Splitting into train and test data set for cross-validation
from sklearn.model_selection import train_test_split

X = scaled_df.iloc[:,:-1]
y = scaled_df['label']

X_train, X_test, y_train, y_test = train_test_split(X.to_numpy(),y.to_numpy(),test_size=0.25,random_state=56)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

!pip install yellowbrick
from yellowbrick.classifier import ConfusionMatrix
from sklearn.model_selection import GridSearchCV,  RepeatedStratifiedKFold
```

```python
#Function to fit and predict values of the model and also fucntions to evaluate model prediciton

def fit_model_prediciton(model,X,y,X_predict):
    model.fit(X,y)
    return model.predict(X_predict)

def accu_score(y_true,y_pred):
    return accuracy_score(y_true,y_pred)

def prec_score(y_true,y_pred):
    return precision_score(y_true,y_pred)

def f_score(y_true,y_pred):
    return f1_score(y_true,y_pred)
```

# Random Forest Classifier

```python
classifier = RandomForestClassifier(max_depth=50,min_samples_split=10,random_state=0)  
```

    ##Train Model Accuracy and Metrics Calcucaltion

```python
y_pred_train = fit_model_prediciton(classifier,X_train,y_train,X_train)
set(y_pred_train)

#Accuracy score
acc = round(accu_score(y_train,y_pred_train),2)
#F1 Score
f1 = round(f_score(y_train,y_pred_train),2)
#Precision score
prec = round(prec_score(y_train,y_pred_train),2)

print(f"Accuracy,F1 and Precision score is {acc}, {f1}, {prec} respectively")

```

```python
cm = ConfusionMatrix(classifier,classes=[0,1])
cm.fit(X_train,y_train)

cm.score(X_train,y_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,y_pred_train))
```
