# Cleaning Reviews

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

cleaning_words = WordNetLemmatizer()

reviews_cl = df.reviews.str.strip("✅ Trip Verified")

new_reviews = []

#Remove punctuations, lowercase,split, lemmatize and join the reviews one by one through for loop

for data in reviews_cl:
    data = re.sub('[^a-zA-Z]',' ', data)
    data = data.lower()
    data = data.split()
    data = [cleaning_words.lemmatize(word) for word in data if word not in set(stopwords.words("english"))]
    data = " ".join(data)
    new_reviews.append(data)
    
```

```python
new_reviews
```

```python
df["Cleaned Reviews"] = new_reviews
df.head(15)
```
